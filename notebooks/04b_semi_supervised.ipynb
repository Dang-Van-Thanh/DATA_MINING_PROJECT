# 04b — Semi-Supervised

```python

import sys
from pathlib import Path
sys.path.append(str(Path().resolve().parent.parent))

import pandas as pd
from sklearn.model_selection import train_test_split
from src.models.semi_supervised import SemiSupervisedLearning
from src.data.cleaner import DataCleaner
```

```python

df=pd.read_csv('data/processed/bank_features.csv')
X=df.drop(columns=['y'])
y=df['y']
```

```python

X_train,X_test,y_train,y_test=train_test_split(
X,y,test_size=0.2,stratify=y,random_state=42)
```

```python

pipe=DataCleaner().create_preprocessing_pipeline()
X_train=pipe.fit_transform(X_train)
X_test=pipe.transform(X_test)
```

## Run Experiment

```python

semi=SemiSupervisedLearning()
res=semi.run_experiment(X_train,y_train,X_test,y_test)
res
```


Insight:
Semi-supervised hữu ích khi nhãn ít → cải thiện performance.
